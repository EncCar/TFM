## 2024

```python
ruta_2024 = os.path.join('carga_2024/data_cargo_2024.xlsx')
df_2024 = pd.read_excel(ruta_2024)
display(df_2024.head(10))
```

```python
df_2024.shape
```

```python
df_anyos = pd.concat([df_2022_2023, df_2024], axis=0)
```

```python
df_anyos
```

```python
df_anyos = df_anyos.rename(columns={'Unnamed: 0':'ID'})
```

```python
df_anyos
```

```python
cond = df_anyos['Meses'].str.contains('2022', case=False)
df_anyos['Anios'] = np.where(cond, '2022', '2023')
```

```python
df_anyos.loc[df_anyos['Meses'].str.contains('2024'), 'Anios'] = '2024'
```

```python
df_anyos['Mes'] = '1'
```

```python
df_anyos.loc[df_anyos['Meses'].str.contains('Febrero'), 'Mes'] = '2'
df_anyos.loc[df_anyos['Meses'].str.contains('Marzo'), 'Mes'] = '3'
df_anyos.loc[df_anyos['Meses'].str.contains('Abril'), 'Mes'] = '4'
df_anyos.loc[df_anyos['Meses'].str.contains('Mayo'), 'Mes'] = '5'
df_anyos.loc[df_anyos['Meses'].str.contains('Junio'), 'Mes'] = '6'
df_anyos.loc[df_anyos['Meses'].str.contains('Julio'), 'Mes'] = '7'
df_anyos.loc[df_anyos['Meses'].str.contains('Agosto'), 'Mes'] = '8'
df_anyos.loc[df_anyos['Meses'].str.contains('Septiembre'), 'Mes'] = '9'
df_anyos.loc[df_anyos['Meses'].str.contains('Octubre'), 'Mes'] = '10'
df_anyos.loc[df_anyos['Meses'].str.contains('Noviembre'), 'Mes'] = '11'
df_anyos.loc[df_anyos['Meses'].str.contains('Diciembre'), 'Mes'] = '12'
```

```python
df.dtypes
```

```python
df_anyos['fecha'] =''+'/'+ df_anyos['Mes']+'/'+df_anyos['Anios']
```

```python
df_anyos['fecha'] = pd.to_datetime(df_anyos['fecha'])
```

```python
# Se calcula el porcentaje de valores faltantes en cada característica.
# % of missing.
for col in df_anyos.columns:
    pct_missing = np.mean(df_anyos[col] == 0)
    print('{} - {}%'.format(col, round(pct_missing*100)))   # las llaves indican las posición de las variables.
```
